```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed()
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N) 
area = (30 * np.random.rand(N))
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(8, 6))
plt.xlabel('Sex')
plt.ylabel('Number')
plt.title('Gender Distribution Bar Chart')
plt.xticks((0, 1), ('Man', 'Woman'))
plt.bar(x=(0, 1), height=(0.75, 0.25), width=0.35, 
        color="red", alpha=0.7, edgecolor='black')
for i, v in enumerate([0.75, 0.25]):
    plt.text(i, v + 0.01, str(v), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(19680801)
N_points = 100000 
n_bins = 20 
x = np.random.randn(N_points)
plt.figure()
ax1 = plt.subplot(111)
ax1.hist(x, bins=n_bins)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
font = {'family': 'DejaVu Sans', 'weight': 'bold', 'size': 14}
matplotlib.rc('font', **font)
x1 = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
x2 = [1, 2, 3, 4, 5]
y2 = [2, 5, 11, 12, 17]
plt.figure(figsize=(10, 6))
plot1 = plt.plot(x1, y1, 'r--*', linewidth=2, markersize=8, label='red line')
plot2 = plt.plot(x2, y2, 'go', markersize=8, label='green circles')

plt.title('Plot of y vs. x')
plt.xticks((1, 2, 3, 4, 5), ('one', 'two', 'three', 'four', 'five'))
plt.xlabel('x axis')   
plt.ylabel('y axis')   
plt.xlim(0.0, 6.0)
plt.ylim(0.0, 30.0)
plt.grid(True, alpha=0.3)
plt.legend(loc='best', shadow=True, ncol=2)
plt.tight_layout()
plt.show()
```
