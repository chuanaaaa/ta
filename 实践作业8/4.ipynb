```python
import numpy as np
X = np.array([[1, 1],
              [1, 2], 
              [1, 3]])
y = np.array([[2],
              [3.1],
              [4]])
β = np.linalg.inv(X.T @ X) @ X.T @ y
print("模型系数:")
print(β)
```
