```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
file_path = r'C:\Users\chuan\Desktop\iris_dataset.csv'
iris_data = pd.read_csv(file_path)
print(iris_data.head())
species = iris_data['target_name'].unique()
print("鸢尾花种类:", species)
```

```python
plt.figure(figsize=(8, 6)) #图形大小
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
species = iris_data['target_name'].unique()
for spec in species:
    subset = iris_data[iris_data['target_name'] == spec]
    plt.scatter(subset['sepal length (cm)'], 
                subset['sepal width (cm)'], 
                c=colors[spec], 
                label=spec,
                alpha=0.7)

plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']  
plt.rcParams['axes.unicode_minus'] = False #使用系统字体
plt.title('鸢尾花数据集二维散点图（花萼长度 vs 花萼宽度）')
plt.xlabel('花萼长度 (cm)')
plt.ylabel('花萼宽度 (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for spec in species:
    subset = iris_data[iris_data['target_name'] == spec]
    ax.scatter(subset['sepal length (cm)'], 
               subset['sepal width (cm)'], 
               subset['petal length (cm)'], 
               c=colors[spec], 
               label=spec,
               alpha=0.7)

ax.set_title('鸢尾花数据集三维散点图')
ax.set_xlabel('花萼长度 (cm)')
ax.set_ylabel('花萼宽度 (cm)')
ax.set_zlabel('花瓣长度 (cm)')
ax.legend()
plt.show()
```
