```python
def my_fan1(A,B):
    m = len(A)      # A的行数
    n = len(B)      # B的行数（等于A的列数）
    p = len(B[0])   # B的列数
    C = [[0] * p for in range(m)]
    for i in range(m):          # 遍历A的每一行
        for j in range(p):      # 遍历B的每一列
            for k in range(n):  # 遍历A的列/B的行
                C[i][j] += A[i][k] * B[k][j]
    return C
    
A = [[1,2]]
B = [[1],[2]]
C =  my_fan1(A,B)
for m in C:
    print(m)
```
