```python
def mean_total_from_list(data_list):
    score_sum = 0
    for score in data_list:
        score_sum = score_sum + score
    mean = score_sum / len(data_list)
    return mean,score_sum
```

```python
score_list = [148,135,120,99,80,82]
score_dict = {"语文":148,"数学":135 ,"外语":120,"物理":99,"生物":80,"化学":82}
```

```python
mean,total = mean_total_from_list(score_list)
print(f"mean = {mean},total = {total}")
```
